```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import pandas as pd
##Q1, 20pts
#%% Q1a, 15pts
# the plot is not required
#Your code is here
print("1a. Write a Python code to find the least square regression line ùë¶ = ùõΩùë• + ùõº. You either")
print("can write a Python code to calculate ùõº and ùõΩ based on the equations above or use")
print("the linear regression function in the model sklearn (described in the lecture) to get")
print("the results directly.")
x_year = np.array([2005,2006,2007,2008,2009])
x1_year = np.array([2005,2006,2007,2008,2009,2012]).reshape((-1,1))
y_sales = np.array([12,19,29,37,45])
# plt.xlabel("X")
# plt.ylabel("Y")
# plt.scatter(x_year,y_sales)
model = linear_model.LinearRegression()
model.fit(x_year.reshape(-1,1),y_sales.reshape(-1,1))
print("ùõº = ", model.intercept_)
print("ùõΩ = ", model.coef_)
print("y = {} + {}X".format(model.intercept_,model.coef_))
print("\n2b. Use the regression line as a model to estimate the sales of the company in 2012 (in")
print("million dollars")
#%%Q1b, 5pts
#Your code is here
#Output the result by print('Estimated sales in 2012 is ', XXX)
y_pred = model.predict(x1_year)
plt.xlabel("Year")
plt.ylabel("Sales(in million dollars)")
plt.title("Sales per Year")
plt.scatter(x_year,y_sales)
plt.plot(x1_year,y_pred, color = 'r')
plt.show()
print("Estimated sales in 2012 is $70 million")
```
