```python
#%% Q2
print("Question 2 **************************************")
print("Pearson correlation")

import matplotlib.pyplot as plt
import numpy as np

x = np.array([50, 68, 74, 70, 65, 61, 63, 74, 62])
y = np.array([170, 193, 209, 185, 195, 188, 188, 202, 183])
plt.scatter(x,y)
plt.title('Scatter Plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
def zscore(arr):
 return(arr - np.mean(arr))/np.std(arr)
 
zx = zscore(x)
zy = zscore(y)

print("zx:")
print(zx)
print("zy:")
print(zy)
plt.scatter(zx,zy)
plt.title('Scatter Plot of zx and zy')
plt.xlabel('zx')
plt.ylabel('zy')
plt.show()
pi = zx.dot(zy)/len(x)
print("Pearson Coefficient:")
print(pi)
# using np.corrcoef gives the same result
nppi = np.corrcoef(x,y)
print("Pearson Coefficient using np.corrcoef:")
print(nppi)
print("WOW they are actually the same")
```

```python
#%% Q3
print("Question 3 **************************************")
# print("P(A U B U C) = P(A) + ... + ...")
print("Pearson correlation")
# pearson correlation coefficient is sensitive to outliers 
# spearman rank coefficient is not sensitive to outliers

import matplotlib.pyplot as plt
import numpy as np

x = np.array([50, 68, 74, 70, 65, 61, 63, 74, 62, 80])
y = np.array([170, 193, 209, 185, 195, 188, 188, 202, 183,1000])
plt.scatter(x,y)
plt.title('Scatter Plot of x and y with outliers')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

def zscore(arr):
 return(arr - np.mean(arr))/np.std(arr)

nppi = np.corrcoef(x,y)
print("Pearson Coefficient using np.corrcoef:")
print(nppi)

x_rank = np.argsort(np.argsort(x))
y_rank = np.argsort(np.argsort(y))

spearman = np.corrcoef(x_rank, y_rank)[0,1]
print("Spearman Correlation Coefficient:")
print(spearman)
print("The two coefficeints are not the same but they are similar")
print("Spearman coefficeint is better because it is closer to 1")
print("0: no correlation,\n1: there is a correlation, \n-1: inverse correlation")
```
