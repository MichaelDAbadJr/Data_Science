```python
import matplotlib.pyplot as plt
#Question 2


# Save the data in the variable "data"
# Each row is saved in a tuple 
#data = [('2016-10-03', 772.559998), ... ]

#2. your code is here
data = [('2016-10-03', 772.55998),('2016-10-04', 776.429993), ('2016-10-05', 776.469971), ('2016-10-06', 776.859985), ('2016-10-07', 775.080017)]
[date, closing] = zip(*data)
plt.plot(date, closing, 'ro-')


# Plot the data as a red line with round markers
# add a title
plt.title("Closing stock value of Alphabet Inc.")

# add a label to the y-axis
plt.ylabel("Closing Value")

# add a label to the x-axis
plt.xlabel("Date")

# Turn on the minor TICKS, which are required for the minor GRID
plt.minorticks_on()

# customize the major grid
plt.grid(which = 'major', linestyle = '-', linewidth = '0.5', color = 'red')
# customize the minor grid
plt.grid(which = 'minor', linestyle = ':', linewidth = '0.5', color = 'black')

plt.show()
```
